```python
import pandas as pd
```

```python
dados_hosp = pd.read_json("dados_hospedagem.json")
dados_hosp.head()
```

```python
dados_hosp = pd.json_normalize(dados_hosp["info_moveis"])
dados_hosp
```

```python
colunas = list(dados_hosp.columns)
colunas
```

```python
dados_hosp = dados_hosp.explode(colunas[3:])
dados_hosp
```

```python
dados_hosp.reset_index(inplace=True,drop=True)
dados_hosp.head()
```

```python
dados_hosp.info()
```

```python
# Convertendo dados numericos:
import numpy as np
```

```python
dados_hosp['max_hospedes'] = dados_hosp['max_hospedes'].astype(np.int64)
dados_hosp.info()
```

```python
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']
```

```python
dados_hosp[col_numericas] = dados_hosp[col_numericas].astype(np.int64)
dados_hosp.info()
```

```python
dados_hosp['avaliacao_geral'] = dados_hosp['avaliacao_geral'].astype(np.float64)
dados_hosp.info()
```

```python
dados_hosp['preco'] = dados_hosp['preco'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

```python
dados_hosp['preco'] = dados_hosp['preco'].astype(np.float64)
```

```python
dados_hosp.info()
```

```python
dados_hosp[['taxa_deposito', 'taxa_limpeza']] = dados_hosp[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x: x.replace('$','').replace(',','').strip())
```

```python
dados_hosp[['taxa_deposito', 'taxa_limpeza']] = dados_hosp[['taxa_deposito', 'taxa_limpeza']].astype(np.float64)
```

```python
dados_hosp.info()
```
